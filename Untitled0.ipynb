```python
import numpy as np
A = np.array([
    [10, 5, -6],
    [2, 2, -6],
    [3, -3, 1]
])
print(A)

B = ([10, -5, -12])
print(B)
invA=np.linalg.inv(A)
X=np.matmul(invA,B)
print(X)
```

```python
print(np.linalg.solve(A,B))
```

```python
#Solve using cramers rule
A1=np.vstack((B,A[:,1],A[:,2]))
A2=np.vstack((A[:,0],B,A[:,2]))
A3=np.vstack((A[:,0],A[:,2],B))
x=np.linalg.det(A1)/np.linalg.det(A)
y=np.linalg.det(A1)/np.linalg.det(A)
z=np.linalg.det(A1)/np.linalg.det(A)
print(x,y,z)
```
